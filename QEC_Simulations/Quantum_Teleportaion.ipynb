```python
qc_alice=QuantumRegister(2 , name = 'alice')
qc_bob=QuantumRegister(1, name ='bob')
alice_reg_1 = ClassicalRegister(1,'bell_1')
alice_reg_2 = ClassicalRegister(1,'bell_2')
cr = ClassicalRegister(3)
qc = QuantumCircuit(qc_alice,qc_bob)
qc.add_register(alice_reg_1)
qc.add_register(alice_reg_2)
#cr=ClassicalRegister(2)
initial_state = [1/np.sqrt(3), np.sqrt(2)/np.sqrt(3)]  
#initial_state = [1, 0]  
qc.h(1)
qc.cx(1,2)
qc.barrier()
qc.initialize(initial_state, 0) 
qc.barrier()
qc.cx(0,1)
qc.h(0)
qc.barrier()
qc.measure(qc_alice[0], alice_reg_1)
qc.measure(qc_alice[1], alice_reg_2)
qc.x(qc_bob[0]).c_if(alice_reg_1, 1)
qc.z(qc_bob[0]).c_if(alice_reg_2, 1)
qc.barrier()
qc.initialize(initial_state, 2) 
qc.draw('mpl')
plt.savefig('teleportation.eps')
plt.show()
```
